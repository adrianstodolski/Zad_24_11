```python
from matplotlib import pyplot as plt


def funkcja_kwadratowa(a, b, c):
    xaxis = []
    yaxis = []
    for x in range(0, 100):
        xaxis.append(x)
        yaxis.append(a*(x**2) + b*x + c)
        
    return xaxis, yaxis

x, y = funkcja_kwadratowa(2, 1, 5)
plt.plot(x, y)
```

```python
import matplotlib.pyplot as plt
import math

def wykres(x, y):
    plt.plot(x, y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Rzut ukosny')
    
def pot(start, koniec, interval):
    numbers = []
    while start < koniec:
        numbers.append(start)
        start = start + interval
        
    return numbers

def kat_rzutu(u, kat):
    
    kat = math.radians(kat)
    g = 9.81
    
    trajektoria = 2*u*math.sin(kat)/g

    intervals = pot(0, trajektoria, 0.001)
    
    x = []
    y = []
    for t in intervals:
        x.append(u*math.cos(kat)*t)
        y.append(u*math.sin(kat)*t-0.5*g*t*t)
        
    print('Czas lotu: {0}'.format(trajektoria)) 
    print('Maksymalna odległość w poziomie: {0}'.format(max(x)))
    print('Maksymalna odległość w pionie: {0}'.format(max(y)))

    wykres(x, y)
    
if __name__ == '__main__':
    num_traj = int(input('Ile rzutów? '))
    for i in range(num_traj):
        try:
            u = float(input('Prędkosc rzutu (m/s): '))
            kat = float(input('Kąt rzutu (w stopniach): '))
        except ValueError:
            print('Błędne dane')
        else:
            kat_rzutu(u, kat)
    
    plt.show()
```

```python
import matplotlib.pyplot as plt

def fibonacci(n):
    ratio = []
    num = []

    a = 0
    b = 1    
    for i in range(1,n):
        c = a + b    
        num.append(i)
        ratio.append(c/b)
        a = b
        b = c
    rysuj(num,ratio)
    
def rysuj(x,y):
    plt.plot(x, y)
    plt.xlabel('Numery ciągu Fibonacciego')
    
    plt.ylabel('Ratio')
    plt.title('Stosunek między kolejnymi numerami Fibonacciego')

if __name__ == '__main__':
    fibonacci(23)
```
